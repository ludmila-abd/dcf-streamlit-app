```python
# Step 1: Import the required libraries
import yfinance as yf
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configure plot style
sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# Step 2: Choose a stock ticker (example: Apple Inc.)
ticker_symbol = "AAPL"

# Create a ticker object
stock = yf.Ticker(ticker_symbol)
```

```python
# Step 3: Get general company information
info = stock.info

# Show some key fields
print("Company:", info.get("longName"))
print("Sector:", info.get("sector"))
print("Market Cap:", info.get("marketCap"))
print("PE Ratio (Trailing):", info.get("trailingPE"))
print("PEG Ratio:", info.get("pegRatio"))
print("Debt to Equity:", info.get("debtToEquity"))
```

```python
# Step 4: Historical market data (daily prices)
history = stock.history(period="5y")

# Show the head of the DataFrame
history.index = history.index.date
history.head()
```

```python
# Step 5: Plot the closing price over time
plt.plot(history.index, history["Close"], label="Closing Price")
plt.title(f"{ticker_symbol} - Closing Price (5 years)")
plt.xlabel("Date")
plt.ylabel("Price (USD)")
plt.legend()
plt.show()
```

```python
# Step 6: Load and display cleaned Income Statement (key metrics only)

# Get income statement and transpose for readability
income_stmt = stock.financials.T

# Select key financial metrics
selected_income = income_stmt[[
    "Total Revenue",
    "Gross Profit",
    "Operating Income",
    "Net Income",
    "EBITDA"
]]

# Convert to billions of USD
selected_income = selected_income / 1e9

# Rename columns for readability
selected_income.columns = [col + " (B USD)" for col in selected_income.columns]

# Clean index to show only the date (no time or timezone)
selected_income.index = selected_income.index.date

# Display the DataFrame
print(f"{ticker_symbol} - Key Income Statement Metrics (in Billion USD):")
display(selected_income)

# Plot the data
selected_income.plot(kind="bar")
plt.title(f"{ticker_symbol} - Key Financial Metrics (Annual)")
plt.ylabel("Billion USD")
plt.xlabel("Fiscal Year End")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Final Step: Company Snapshot Summary

# Extract info from yfinance
summary = {
    "Company": info.get("longName"),
    "Ticker": ticker_symbol,
    "Sector": info.get("sector"),
    "Industry": info.get("industry"),
    "Market Cap (B USD)": round(info.get("marketCap", 0) / 1e9, 2),
    "Trailing P/E": info.get("trailingPE"),
    "Forward P/E": info.get("forwardPE"),
    "PEG Ratio": info.get("pegRatio"),
    "Debt to Equity": info.get("debtToEquity"),
    "Return on Equity (%)": round(info.get("returnOnEquity", 0) * 100, 2),
    "Revenue (TTM, B USD)": round(info.get("totalRevenue", 0) / 1e9, 2),
    "Net Income (TTM, B USD)": round(info.get("netIncomeToCommon", 0) / 1e9, 2),
    "52-Week Change (%)": round(info.get("52WeekChange", 0) * 100, 2)
}

# Convert to DataFrame for cleaner display
snapshot_df = pd.DataFrame.from_dict(summary, orient='index', columns=["Value"])
display(snapshot_df)
```
